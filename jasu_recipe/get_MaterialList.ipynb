```python
# -＊- coding: utf-8 -＊-

import requests
from pprint import pprint
from time import sleep
from itertools import islice

url = 'https://app.rakuten.co.jp/services/api/Recipe/CategoryRanking/20170426?'

def get_ingredient(id):
    payload = {
       'applicationId': [1098435242683481258],
       'categoryId': id.strip(),
       }
    r = requests.get(url, params=payload)
    resp = r.json()
    for recipe in resp['result']:
        for food in recipe["recipeMaterial"]:
            normalized_food = food.strip(rm_char).replace('(', ' ').replace('（', ' ').split(" ")[0]
            yield normalized_food

lack_set = set()
lack_list = []
idlist_all = open('idlist_all.txt','r')

for Id in islice(idlist_all, 43):
    for j in get_ingredient(Id):
        lack_set.add(j)
    sleep(1.5)
lack_list = list(lack_set)
#print(lack_list)

file = open('ingredient_list.txt', 'w')
for k in lack_list:
    file.writelines(k)
    file.writelines('\n')
    print(k)
    print('\n')
```
